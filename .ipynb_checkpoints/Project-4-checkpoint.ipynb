```python
import pandas as pd
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 120)
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data15 = pd.read_excel('../../../Downloads/archive (1)/Career Mode player datasets - FIFA 15-22.xlsx', sheet_name = 'FIFA 15')
data16 = pd.read_excel('../../../Downloads/archive (1)/Career Mode player datasets - FIFA 15-22.xlsx', sheet_name = 'FIFA 16')
data17 = pd.read_excel('../../../Downloads/archive (1)/Career Mode player datasets - FIFA 15-22.xlsx', sheet_name = 'FIFA 17')
data18 = pd.read_excel('../../../Downloads/archive (1)/Career Mode player datasets - FIFA 15-22.xlsx', sheet_name = 'FIFA 18')
data19 = pd.read_excel('../../../Downloads/archive (1)/Career Mode player datasets - FIFA 15-22.xlsx', sheet_name = 'FIFA 19')
data20 = pd.read_excel('../../../Downloads/archive (1)/Career Mode player datasets - FIFA 15-22.xlsx', sheet_name = 'FIFA 20')
data21 = pd.read_excel('../../../Downloads/archive (1)/Career Mode player datasets - FIFA 15-22.xlsx', sheet_name = 'FIFA 21')
data22 = pd.read_excel('../../../Downloads/archive (1)/Career Mode player datasets - FIFA 15-22.xlsx', sheet_name = 'FIFA 22')
```

```python
data15['Year'] = 15
data16['Year'] = 16
data17['Year'] = 17
data18['Year'] = 18
data19['Year'] = 19
data20['Year'] = 20
data21['Year'] = 21
data22['Year'] = 22
```

```python
data = pd.concat([data15, data16, data17, data18, data19, data20, data21, data22], axis = 0)
```

```python
data.reset_index(inplace = True)
```

```python
data.loc[data.loc[:,'short_name']=='L. Messi']
```

```python
data.drop(columns='index', inplace= True)
```

```python
data.head()
```

```python
top_player_2022 = data.loc[(data.loc[:,'Year'] == 22) & (data.loc[:,'overall'] > 85)]
```

```python
top_player_2022
```

```python
goalkeepers = top_player_2022[top_player_2022.loc[:,'player_positions'] == 'GK'].index
goalkeepers
```

```python
top_player_2022.drop(index = goalkeepers, axis=0, inplace= True)
```

```python
corr = top_player_2022[['overall', 'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic', 'attacking_crossing',
                 'attacking_finishing', 'attacking_heading_accuracy', 'attacking_short_passing','attacking_volleys', 
                 'skill_dribbling', 'skill_curve','skill_fk_accuracy', 'skill_long_passing', 'skill_ball_control',
                 'movement_acceleration', 'movement_sprint_speed', 'movement_agility','movement_reactions', 
                 'movement_balance', 'power_shot_power', 'power_jumping', 'power_stamina', 'power_strength', 
                 'power_long_shots','mentality_aggression', 'mentality_interceptions','mentality_positioning', 
                 'mentality_vision', 'mentality_penalties','mentality_composure', 'defending_marking_awareness']].corr()
```

```python
corr.columns
```

```python
corr.drop(columns = ['pace', 'shooting', 'passing', 'dribbling', 'defending',
       'physic', 'attacking_crossing', 'attacking_finishing',
       'attacking_heading_accuracy', 'attacking_short_passing',
       'attacking_volleys', 'skill_dribbling', 'skill_curve',
       'skill_fk_accuracy', 'skill_long_passing', 'skill_ball_control',
       'movement_acceleration', 'movement_sprint_speed', 'movement_agility',
       'movement_reactions', 'movement_balance', 'power_shot_power',
       'power_jumping', 'power_stamina', 'power_strength', 'power_long_shots',
       'mentality_aggression', 'mentality_interceptions',
       'mentality_positioning', 'mentality_vision', 'mentality_penalties',
       'mentality_composure', 'defending_marking_awareness'], inplace=True)
```

```python
corr.sort_values(by = 'overall', ascending = False)
```

```python
top_players = top_player_2022.groupby(top_player_2022['overall']).agg({'movement_reactions': 'mean', 'mentality_composure':'mean', 'shooting':'mean', 
                                   'power_shot_power':'mean', 'power_long_shots':'mean', 'attacking_finishing':'mean', 
                                  'attacking_volleys':'mean'})
top_players.reset_index(inplace=True)
melt= pd.melt(top_players, id_vars='overall')
sns.barplot(x='overall', y='value', hue='variable',data=melt);
sns.set(rc={'figure.figsize':(20,200)})
```

```python
fig, ax = plt.subplots(2,2,figsize=(20,10))
sns.lineplot(x=top_player_2022['overall'], y=top_player_2022['defending_marking_awareness'], ax=ax[0,0])
sns.lineplot(x=top_player_2022['overall'], y=top_player_2022['defending'], ax=ax[0,1])
sns.lineplot(x=top_player_2022['overall'], y=top_player_2022['mentality_interceptions'], ax=ax[1,0])
sns.lineplot(x=top_player_2022['overall'], y=top_player_2022['mentality_aggression'], ax=ax[1,1]);
```

```python
fig, ax = plt.subplots(2,2,figsize=(20,10))
sns.lineplot(x=top_player_2022['overall'], y=top_player_2022['movement_reactions'], ax=ax[0,0])
sns.lineplot(x=top_player_2022['overall'], y=top_player_2022['mentality_composure'], ax=ax[0,1])
sns.lineplot(x=top_player_2022['overall'], y=top_player_2022['power_shot_power'], ax=ax[1,0])
sns.lineplot(x=top_player_2022['overall'], y=top_player_2022['shooting'], ax=ax[1,1]);
```

```python
sns.barplot(x=top_player_2022['overall'], y=top_player_2022['movement_reactions'])
sns.barplot(x=top_player_2022['overall'], y=top_player_2022['mentality_composure'])
sns.barplot(x=top_player_2022['overall'], y=top_player_2022['power_shot_power'])
sns.barplot(x=top_player_2022['overall'], y=top_player_2022['shooting']);
```
