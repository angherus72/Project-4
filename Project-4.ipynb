```python
def largest(x):
    top_values = []
    attributes = []
    lst = x.columns
    for i in lst:
        top_values.append(x[i].max())
        indexi = x[i].nlargest(1).index.astype(str, copy=False)
        attributes.append(indexi)
    return top_values, attributes
values, attr = largest(corr_fifa)

 
def to_string(y):
    indexes = []
    for i in y:
        indexes.append(str(i))
    return indexes

new = zip(attr, values)
df_new_top = pd.DataFrame(new)
df_new_top.set_index([['dribbling', 'pace', 'pace','shooting','movement_reactions', 
                      'movement_reactions', 'shooting', 'movement_reactions']], inplace = True)
df_new_top.reset_index(inplace=True)
df_new_top.drop(columns=0, inplace=True)
df_new_top['Fifa Year'] = ['FIFA 15', 'FIFA 16', 'FIFA 17', 'FIFA 18', 'FIFA 19', 'FIFA 20', 'FIFA 21', 'FIFA 22']
df_new_top.to_excel('Top_Attribute_Year.xlsx')
```

```python
def avg_sal(x):
    average = []
    for i in x:
        p = i['wage_eur'].sum()
        average.append(p)
    return average
```
